```python
X = df.drop(columns = 'Outcome', axis=1)
Y = df['Outcome']
```

```python
print('X',X)
```

```python
print('Y',Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, stratify=Y, random_state=2)
```

```python
from sklearn.svm import SVC
SVC_i = SVC()
SVC_i.fit(X_train, Y_train)
```

```python
X_train1 = SVC_i.predict(X_test)
training = accuracy_score(X_train1, Y_test)
training
```

```python
X_train2 = SVC_i.predict(X_test)
training = accuracy_score(X_train2, Y_test)
training
```

```python
# accuracy score on the training data
X_train_prediction = SVC_i.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
training_data_accuracy
```

```python
from sklearn.metrics import confusion_matrix, classification_report
y_predictions = SVC_i.predict(X_test)
print(confusion_matrix(Y_test, y_predictions))
```

```python
print(classification_report(Y_test, y_predictions))
```

```python
input_data = (5,166,72,19,175,25.8,0.587,51)

# changing the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = SVC_i.predict(input_data_reshaped)
print(prediction)

if (prediction[0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
