```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/vgsales.csv')
df.head()
```

```python
#Delete the Rank column
df = df.drop('Rank', axis=1)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#Delete null values from data
df.dropna(inplace=True)
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
cols = ['Name', 'Platform', 'Genre',
       'Publisher']
df[cols] = df[cols].apply(LabelEncoder().fit_transform)
df.head()
```

***Decision Tree Regressor***

```python
df1 = df
```

```python
# Importing the libraries
from sklearn.model_selection import train_test_split , GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
# Split the data into features (X) and target variable (y)
X = df1.drop('Global_Sales', axis=1)
y = df1['Global_Sales']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)

# Create an instance of the DecisionTreeRegressor
model = DecisionTreeRegressor()

# Fit the model to the training data
model.fit(X_train, y_train)

# Predict the target variable for the testing set
y_pred = model.predict(X_test)

# Evaluate the model using mean squared error and r2_score
mse = mean_squared_error(y_test, y_pred)
r_square = r2_score(y_test, y_pred)
print('R-squared:', r_square)
print('Mean Squared Error:', mse)
```

```python
model.score(X_test, y_test)
```

***Random Forest Regression***

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
mse1 = mean_squared_error(y_test, y_pred)
r_square1 = r2_score(y_test, y_pred)
print('R-squared:', r_square1)
print('Mean Squared Error:', mse1)
```

```python
regressor.score(X_test, y_test)
```

```python
regressor2 = RandomForestRegressor(n_estimators = 50, random_state = 0,max_depth=10)
regressor2.fit(X_train, y_train)
y_pred2 = regressor2.predict(X_test)
mse3 = mean_squared_error(y_test, y_pred2)
r_square3 = r2_score(y_test, y_pred2)
print('R-squared:', r_square3)
print('Mean Squared Error:', mse3)
```

***Grid Search CV***

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
# determine the best value of the RF hyperparameter
clf = GridSearchCV(RandomForestRegressor(), {
    'n_estimators': [10,30],
    'max_depth': [1,2,10],
}, cv=2, return_train_score=False)
clf.fit(X_train,y_train)
```

```python
rfCV = pd.DataFrame(clf.cv_results_)
rfCV
```

```python
rfCV[['param_n_estimators','params','mean_test_score']]
```

```python
clf.best_params_
```

***Randomized Search CV***

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
clf1 = RandomizedSearchCV(RandomForestRegressor(), {
    'n_estimators': [10,30],
    'max_depth': [1,2,10],
}, cv=2, return_train_score=False)
clf1.fit(X_train,y_train)
```

```python
clf1.best_params_
```

```python
RmCV = pd.DataFrame(clf1.cv_results_)
RmCV[['param_n_estimators','params','mean_test_score']]
```

```python
clf1.score(X_test, y_test)
```
